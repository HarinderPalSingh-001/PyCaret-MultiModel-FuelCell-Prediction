```python
#Make prediction on "new unseen dataset"
!pip install pycaret

import pandas as pd
from pycaret.regression import *

# Step 1: Load your dataset using pandas (if not already done)
bostonDataSet = pd.read_csv("/content/Fuel_cell_performance_data-Full.csv")

# Step 2: Setup PyCaret environment
setup(data=bostonDataSet, target='Target5', train_size=0.7, data_split_shuffle=False, verbose=False)

# Step 3: Train a Random Forest model (assuming 'rfModel' is trained)
rfModel = create_model('rf')  # 'rf' stands for Random Forest

# Step 4: Make predictions using the trained model on the new data
newDataSet = bostonDataSet.iloc[:10]  # Select a subset of the dataset for predictions
newPredictions = predict_model(rfModel, data=newDataSet)

# Step 5: Display the predictions
print(newPredictions)
```

```python
#"Scatter plot" b/w actual and predicted
import matplotlib.pyplot as plt

predicted = newPredictions.iloc[:,-1]     # Last column
actual = newPredictions.iloc[:,-2]        # 2nd last column

plt.scatter(actual, predicted)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Actul Vs Predicted')
plt.savefig("result-scatter-plot.jpg", dpi=300)
plt.show()
```
