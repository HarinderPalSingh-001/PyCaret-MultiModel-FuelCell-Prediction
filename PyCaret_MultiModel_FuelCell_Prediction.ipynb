```python
#Model performance using "Outlier Removal"


from pycaret.regression import setup, compare_models

# Prepare the dataset with outlier removal using Isolation Forest and set up the regression environment
data_with_outliers_removed = bostonDataSet.copy()  # Ensure the original dataset remains unchanged
regression_setup = setup(data=data_with_outliers_removed, target='Target5',
                         remove_outliers=True, outliers_method='iforest', outliers_threshold=0.05,
                         data_split_shuffle=False, verbose=False)

# Compare different regression models after performing outlier removal
model_comparison_result = compare_models()
```

```python
#Model performance using "PCA"
from pycaret.regression import setup, compare_models

# Prepare the dataset for regression, applying Principal Component Analysis (PCA) with a linear method
pca_data = bostonDataSet.copy()  # Ensure original dataset remains unmodified
pca_setup = setup(data=pca_data, target='Target5',
                  pca=True, pca_method='linear', data_split_shuffle=False, verbose=False)

# Compare multiple regression models after applying PCA
pca_model_comparison = compare_models()
```
